```python
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "ai"],
    ["ai", "is", "powerful"],
    ["i", "love", "machine", "learning"]
]

model = Word2Vec(
    sentences=sentences,
    vector_size=50,
    window=2,
    sg=1,          # 1 = Skip-gram
    min_count=1    # important for small dataset
)

# Check vector of a word
print(model.wv["ai"])
```
