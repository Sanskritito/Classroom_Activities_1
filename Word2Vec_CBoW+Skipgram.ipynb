```python
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "deep", "learning"],
    ["deep", "learning", "is", "powerful"],
    ["word2vec", "creates", "embeddings"],
    ["i", "love", "nlp"]
]

# CBOW
cbow_model = Word2Vec(sentences,
                      vector_size=100,
                      window=3,
                      min_count=1,
                      sg=0)   # sg=0 → CBOW

# Skip-gram
skip_model = Word2Vec(sentences,
                      vector_size=100,
                      window=3,
                      min_count=1,
                      sg=1)   # sg=1 → Skip-gram

print("CBOW similar to love:", cbow_model.wv.most_similar("love"))
print("Skipgram similar to love:", skip_model.wv.most_similar("love"))
```
